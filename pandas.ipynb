```python
import pandas as pd
```

```python
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
```

```python
# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
```

```python
df.head() # First 5 rows
df.tail() # Last 5 rows
df.info() # Info about DataFrame
df.describe() # Summary statistics
```

```python
# Column selection
df['Name']
df[['Name','Age']]
```

```python
# Row selection by label and position

df.loc[0]
df.iloc[1]
```

```python
df[df['Age'] > 25]
```

```python
df['Gender'] = ['F','M'] # Add new column
```

```python
df.isnull()
df.dropna()
df.fillna(0)
```

```python
df.groupby('Gender')['Age'].mean()
# df.drop('Age', axis=1, inplace=True)
```

```python
df1 = pd.DataFrame({
    'key': [1, 2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'key': [1, 2],
    'Age': [25, 30]
})

pd.merge(df1, df2, on='key')
```
