```python

#To check wether a number is even or odd
num = int(input("Enter a number:"))
if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
#To create a grade program
mark = int(input("Enter mark:"))
if mark >= 90:
    print("A Grade")
elif mark >= 75:
    print("B Grade")
elif mark >= 50:
    print("C Grade")
else:
    print("Fail")
    
```
