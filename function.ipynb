```python
# Function to return the square of a number
def square(num):
    return num * num
print(square(5))
```

```python
# Function that adds two numbers
def add(a,b):
    return a + b
print(add(10,20))
```

```python
# Function to check if the number is even
def is_even(num):
    return num % 2 == 0
print(is_even(8))
```

```python
# Function that returns the largest number from a list
def largest(lst):
    return max(lst)
print(largest([10,45,23,89,12]))
```

```python
# Function to compute factorial of a number
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
print(factorial(5))
```

```python
#Function to reverse a string
def reverse_string(text):
    return text[::-1]
print(reverse_string("Python"))
```

```python
#function to check if a string a palindrome 
def palidrome(text):
    return text == text[::-1]

print(palidrome("madam"))
```

```python
#function to find the sum of elements in a list
def list_sum(lst):
    return sum(lst)

print(list_sum([1 ,2 ,3 ,4 ,5 ]))

```

```python
# Function to count vowels in a string
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in text:
        if ch in vowels:
            count += 1
    return count

print(count_vowels("Hello World"))
```

```python
# Function that checks if a number is prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print(is_prime(7))
```

```python
#Function to generate Fibonacci series up to n terms
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)
print()
```

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(celsius_to_fahrenheit(25))
```

```python
# Function to calculate the area of a circle
def area_circle(radius):
    return 3.14 * radius * radius

print(area_circle(7))
```

```python
# Function that returns the minimum number from a list
def minimum(lst):
    return min(lst)

print(minimum([4, 8, 1, 9, 2]))
```

```python
#Function to concatenate two strings
def concatenate(str1, str2):
    return str1 + str2

print(concatenate("Hello ", "World"))
```

```python
# Function that finds the GCD of two numbers
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(12, 18))
```

```python
#Function that accepts a list and returns it in sorted order
def sort_list(lst):
    return sorted(lst)

print(sort_list([5, 2, 8, 1, 3]))
```

```python
# Function to count the number of words in a string
def word_count(text):
    words = text.split()
    return len(words)

print(word_count("Python is easy to learn"))
```

```python
#Function that accepts a number and returns the sum of its digits
def digit_sum(num):
    total = 0
    while num > 0:
        total += num % 10
        num //= 10
    return total

print(digit_sum(1234))
```

```python
#Function to check if a given year is a leap year
def leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False

print(leap_year(2024))
```
