Reading from a file

```python
f = open("example.txt", "r") # Open file in read mode
content = f.read() # Read entire file
print(content)
f.close()
```

Writing to a file


```python
f = open("example.txt", "w") #opens the file in write mode
f.write("Hello, jobin,file Handling!") 
f.close() 
```

Appending to a file

```python
f = open("example.txt", "a") #open file in append mode
f.write("\nThis is an additional line.")
f.close()
```

```python
try:
    num = int("abc") # This will raise a ValueError
              
except ValueError:
    print("Invalid input! Please enter a number.")
else:
    print("No Error")
```

Catching Specific Exception Type

```python
try:
    result = int("100a")
    #result = 10 / 0

except ZeroDivisionError:
    print("Division by zero is not allowed.")
#except ValueError:
#   print("Invalid value.")
except Exception as e: # Catching any other unexpected exceptions
    print(f"An unexpected error occurred: {e}")
```

finally Statement

```python
try:
    file = open("sample.txt", "w")
    file.write("Hello")
    #print(1 / 0) # Will raise ZeroDivisionError
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print("No Error")
finally:
    file.close()
    print("File closed.")
```

Combined Example

```python
try:
    num = int(input("Enter a number: "))
    assert num != 0, "Number cannot be zero."
    result = 10 / num
except ValueError:
    print("Please enter a valid integer.")
except AssertionError as e:
    print(e)
except ZeroDivisionError:
    print("Division by zero is not allowed.")
finally:
    print("Execution completed.")
```
